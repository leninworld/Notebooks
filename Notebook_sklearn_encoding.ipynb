```python
def pretty_print(title, df, n):
  print(f"\n {title}")
  print(tabulate(df.head(n),headers='keys',tablefmt='psql'))
```

```python
# read titanic 
df_in_titanic = pd.read_csv(path_titanic)
# read powerlifting
df_in_power = pd.read_csv(path_power)
```

```python
# LABEL ENCODER EXAMPLE WITH POWERLIFTING DATA SET

labelencoder = LabelEncoder()
df_usa_men_val['encoded_equipment'] = labelencoder.fit_transform(df_usa_men_val['Equipment'])

pretty_print("title ", df_usa_men_val, 2)

lst_sel_cols = ["Equipment", "encoded_equipment"]

pretty_print("Selective columns print Equipment and encoded_equipment:", df_usa_men_val[lst_sel_cols], 10)
```

```python
# ONE HOT ENCODING EXAMPLE WITH POWERLIFTING DATA SET ("SEX")

# usa men only
df_usa = df_in_power[(df_in_power["Country"] == "USA")]
# print distinct gender
print(f"distinct gender -> {df_in_power['Sex'].unique()}")
# drop na for "Sex"
# df_usa["Sex"].dropna(inplace=True)
df_usa.dropna(subset=['Sex'], how='all', inplace=True)
# age with valid numbers only (above 10)
df_usa_val = df_usa[df_usa["Age"] > 10]
# one hot encoder
enc = OneHotEncoder(handle_unknown='ignore')
# pass "Sex" column for fit into transformation
enc_df = pd.DataFrame(enc.fit_transform(df_usa_val[['Sex']]).toarray())
# merge both df
df_merge = df_usa_val.join(enc_df)
# list of new merged df (three new columns since three distinct value for column "Sex" 
# as "M", "F", "Mx")
lst_columns = list(df_usa_val) + ["M", "F", "O"]
# rename the new columns
df_merge.columns = lst_columns
# three new columns (corresponding to "M", "F", "Mx" in column "Sex")
lst_sel_cols = ["M", "F", "O"]
lst_sel_cols_2 = ["Sex"] + lst_sel_cols

# "M" only
filter_1 = df_merge["M"] == 1
pretty_print("Sex: 'M' only", df_merge[filter_1][lst_sel_cols_2], 10)
# "F" only
filter_2 = df_merge["F"] == 1
pretty_print("Sex: 'F' only", df_merge[filter_2][lst_sel_cols_2], 10)
# "Mx" only
filter_3 = df_merge["Sex"] == "Mx"
pretty_print("Sex: 'O' only", df_merge[filter_3][lst_sel_cols_2], 10)
```

```python
# LABEL ENCODER EXAMPLE WITH TITANIC DATA SET

# actual read from the csv file
pretty_print("titanic:", df_in_titanic, 2)

# take only categorical data with select_dtypes
df_in_titanic_cat = df_in_titanic.select_dtypes(include=[object])
pretty_print("titanic (cat only):", df_in_titanic_cat, 2)
# label encoder
le = LabelEncoder()
# fit transform
df_tit_label = df_in_titanic_cat.apply(le.fit_transform)
pretty_print("label encoded for categorical columns only:", df_tit_label, 5)
```
